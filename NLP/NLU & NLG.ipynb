```python
import spacy
from collections import Counter
from string import punctuation
```

```python
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
def get_hotwords(text):
    result =[]
    pos_tag =['PROPN','ADJ','NOUN']
    doc= nlp(text.lower())
    for token in doc:
        if (token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
            if(token.pos_ in pos_tag):
                result.append(token.text)
    return result
```

```python
aurora="An aurora pl. aurorae or auroras is a natural light display in Earths sky, predominantly observed in high-latitude regions around the Arctic and Antarctic. The terms northern lights (aurora borealis) and southern lights (aurora australis) are used in the Northern and Southern Hemispheres respectively.[3][c] Auroras display dynamic patterns of radiant light that appear as curtains, rays, spirals or dynamic flickers covering the entire sky"
```

```python
get_hotwords(aurora)
```

# Step3 = Stemming/ lemmatization

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
ps = PorterStemmer()
words = ['jumped','daily','better','eating','captured','building']
```

```python
for word in words:
    print(f"{word} => {ps.stem(word)}")
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemma=WordNetLemmatizer()
for word in words:
    print(f"{word}=>{lemma.lemmatize(word,pos='v')}")
```

# stemming vs Lemmatization
* stemming is faster than lemmatization
* stemming chops of words irrespective of context
* lemmatizer is context dependent has parts of speech relevancy dictonary based approach
* lemmatizer is more accurate than stemming

```python
sent=['This is the first piece of document loaded.',
     'this is second of same kind.',
     'this is third document',
     'whats this kind of then']
print(sent)
```

```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v= CountVectorizer()
X=v.fit_transform(sent)
columns=v.get_feature_names_out()
df=pd.DataFrame(X.toarray(),columns=columns,index=sent)
df
```

```python
print(v.vocabulary_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
Y=tfidf.fit_transform(sent)
df_tfidf = pd.DataFrame(Y.toarray(),columns=tfidf.get_feature_names_out(),index=sent)
df_tfidf
```

* Emphasis is on important words
* Doesn't capture word similarity/context
* for a huge sentence results in sparse matrix
