```python
# to avoid overfitting there are two methods  1)early stopping
model=Sequential()
model.add(InputLayer(shape=(30,))) #input layer with 30 ip features
model.add(Dense(26,activation='relu')) #1st hidden layer
model.add(Dense(15,activation='relu')) #2nd hidden layer
model.add(Dense(1,activation='sigmoid'))# output layer
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping 
early_stop = EarlyStopping(monitor='val_loss',mode='min',patience=50)
early_stop
```

```python
model.fit(x=X_train,y=y_train,epochs=100,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
m_loss=pd.DataFrame(model.history.history)
m_loss.plot()
```

```python
# Adding Dropout Layers

from tensorflow.keras.layers import Dropout
```

```python
model= Sequential()
model.add(InputLayer(shape=(30,))) #input layer with 30 ip features
model.add(Dense(26,activation='relu')) #1st hidden layer
model.add(Dropout(0.5))   #percentage of the connections to be cutoff dropout(it should not exceed 60%)
model.add(Dense(15,activation='relu')) #2nd hidden layer
model.add(Dropout(0.5)) 
model.add(Dense(1,activation='sigmoid'))# output layer
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```
