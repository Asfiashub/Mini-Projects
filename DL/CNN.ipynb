# ANN MLP- Multi layer perceptron

```python

import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
cifar10_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
from tensorflow import keras
from keras.layers import Dense,Flatten
```

```python
ann=keras.Sequential()
ann.add(Flatten(input_shape=(32,32,3))) 
ann.add(Dense(2048,activation='relu'))
ann.add(Dense(10,activation='softmax'))
```

```python
ann.summary()
```

```python
ann.compile(loss='sparse_categorical_crossentropy',
           optimizer='adam',metrics=['accuracy'])
```

```python
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
model_ann= ann.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
print(plt.plot(model_ann.history['loss']))
plt.plot(model_ann.history['val_loss'])
```

# CNN-convolution NN 

dataset used here is cifar

```python
from tensorflow import keras
from keras.layers import Conv2D,Dense,Flatten,MaxPooling2D,Dropout
```

```python
from tensorflow import keras
cnn= keras.Sequential()
cnn.add(Conv2D(32,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu',input_shape=(32,32,3)))
cnn.add(MaxPooling2D((2,2)))                    # pooling layer
cnn.add(Conv2D(64,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')) # 1st hidden layer edges&lines
cnn.add(MaxPooling2D((2,2)))
cnn.add(Conv2D(128,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')) # 2nd- shapes,curves
cnn.add(MaxPooling2D((2,2)))
cnn.add(Conv2D(256,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')) # 3rd objects and parts
cnn.add(MaxPooling2D((2,2)))
cnn.add(Flatten())
cnn.add(Dense(64,activation='relu')) 
cnn.add(Dropout(0.3))
cnn.add(Dense(10,activation='softmax')) #output layer
```

```python
cnn.summary()
```

```python
cnn.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
history = cnn.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
def Accuracy_val(history):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
Accuracy_val(history)
```

```python
def plotLosses(history):
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper right')
    plt.show()
```

```python
plotLosses(history)
```
